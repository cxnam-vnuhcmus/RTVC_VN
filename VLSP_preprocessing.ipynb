```python
from vinorm import TTSnorm
new_transcript = []
for line in open('datasets_fpt/fpt/transcript_all.txt', 'rt').readlines():
    parts = line.strip().split('\t')
    text = TTSnorm(parts[1])
    text = text.replace('.', '')
    new_transcript.append('{}\t{}'.format(parts[0], text))

with open('./transcript_all_new.txt', 'wt') as f:
    for t in new_transcript:
        f.write(t + '\n')
```
